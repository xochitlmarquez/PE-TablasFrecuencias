```python
import random
import pandas as pd

# Contexto: Encuesta sobre frutas favoritas
frutas = [
    "Manzana", "Pl치tano", "Naranja", "Fresa", 
    "Uva", "Kiwi", "Mango", "Pi침a", "Papaya"
]

# Generar 500 respuestas aleatorias
respuestas = random.choices(frutas, k=500)

# Crear un DataFrame para contar las frecuencias
frecuencia = pd.Series(respuestas).value_counts().reset_index()
frecuencia.columns = ['Fruta', 'Frecuencia Absoluta']

# Calcular la frecuencia relativa
frecuencia['Frecuencia Relativa'] = frecuencia['Frecuencia Absoluta'] / 500

# Calcular la frecuencia acumulada
frecuencia['Frecuencia Acumulada'] = frecuencia['Frecuencia Absoluta'].cumsum()

# Ordenar por la fruta
frecuencia = frecuencia.sort_values(by='Fruta').reset_index(drop=True)

# Mostrar la tabla de frecuencia
print(frecuencia)

# Guardar la tabla en un archivo CSV (opcional)
frecuencia.to_csv('tabla_frecuencia_frutas.csv', index=False)
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Contexto: Encuesta sobre frutas favoritas
frutas = [
    "Manzana", "Pl치tano", "Naranja", "Fresa", 
    "Uva", "Kiwi", "Mango", "Pi침a", "Papaya"
]

# Generar 500 respuestas aleatorias
respuestas = random.choices(frutas, k=500)

# Crear un DataFrame para contar las frecuencias
frecuencia = pd.Series(respuestas).value_counts().reset_index()
frecuencia.columns = ['Fruta', 'Frecuencia Absoluta']

# Calcular la frecuencia relativa
frecuencia['Frecuencia Relativa'] = frecuencia['Frecuencia Absoluta'] / 500

# Calcular la frecuencia acumulada
frecuencia['Frecuencia Acumulada'] = frecuencia['Frecuencia Absoluta'].cumsum()

# Ordenar de mayor a menor
frecuencia = frecuencia.sort_values(by='Frecuencia Absoluta', ascending=False).reset_index(drop=True)

# Colores para las barras
colores = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'cyan', 'magenta']

# Gr치fica 1: Histograma de Frecuencia Absoluta
plt.figure(figsize=(10, 6))
plt.bar(frecuencia['Fruta'], frecuencia['Frecuencia Absoluta'], color=colores[:len(frecuencia)], width=0.6)
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gr치fica 2: Diagrama de Barras Horizontales de Frecuencia Absoluta
plt.figure(figsize=(10, 6))
plt.barh(frecuencia['Fruta'][::-1], frecuencia['Frecuencia Absoluta'][::-1], color=colores[:len(frecuencia)])
plt.title('Diagrama de Barras Horizontales de Frecuencia Absoluta')
plt.xlabel('Frecuencia Absoluta')
plt.ylabel('Fruta')
plt.tight_layout()
plt.show()

# Gr치fica 3: Pol칤gono de Frecuencia
plt.figure(figsize=(10, 6))
plt.plot(frecuencia['Fruta'], frecuencia['Frecuencia Absoluta'], marker='o')
plt.title('Pol칤gono de Frecuencia')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.grid(True)
plt.fill_between(frecuencia['Fruta'], 0, frecuencia['Frecuencia Absoluta'], alpha=0.1)
plt.tight_layout()
plt.show()

# Gr치fica 4: Diagrama de Pastel de Frecuencia Relativa
plt.figure(figsize=(8, 8))
plt.pie(frecuencia['Frecuencia Relativa'], labels=frecuencia['Fruta'], autopct='%1.1f%%', startangle=90)
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Gr치fica 5: Ojiva de Frecuencia Acumulada
plt.figure(figsize=(10, 6))
plt.plot(frecuencia['Fruta'], frecuencia['Frecuencia Acumulada'], marker='o', color='orange')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.fill_between(frecuencia['Fruta'], 0, frecuencia['Frecuencia Acumulada'], alpha=0.1)
plt.show()
```
