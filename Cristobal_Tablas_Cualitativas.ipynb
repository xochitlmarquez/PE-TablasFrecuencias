```python
import random
import pandas as pd

# Contexto: Encuesta sobre frutas favoritas
frutas = [
    "Manzana", "Plátano", "Naranja", "Fresa", 
    "Uva", "Kiwi", "Mango", "Piña", "Papaya"
]

# Generar 500 respuestas aleatorias
respuestas = random.choices(frutas, k=500)

# Crear un DataFrame para contar las frecuencias
frecuencia = pd.Series(respuestas).value_counts().reset_index()
frecuencia.columns = ['Fruta', 'Frecuencia Absoluta']

# Calcular la frecuencia relativa
frecuencia['Frecuencia Relativa'] = frecuencia['Frecuencia Absoluta'] / 500

# Calcular la frecuencia acumulada
frecuencia['Frecuencia Acumulada'] = frecuencia['Frecuencia Absoluta'].cumsum()

# Ordenar por la fruta
frecuencia = frecuencia.sort_values(by='Fruta').reset_index(drop=True)

# Mostrar la tabla de frecuencia
print(frecuencia)

# Guardar la tabla en un archivo CSV (opcional)
frecuencia.to_csv('tabla_frecuencia_frutas.csv', index=False)
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Contexto: Encuesta sobre frutas favoritas
frutas = [
    "Manzana", "Plátano", "Naranja", "Fresa", 
    "Uva", "Kiwi", "Mango", "Piña", "Papaya"
]

# Generar 500 respuestas aleatorias
respuestas = random.choices(frutas, k=500)

# Crear un DataFrame para contar las frecuencias
frecuencia = pd.Series(respuestas).value_counts().reset_index()
frecuencia.columns = ['Fruta', 'Frecuencia Absoluta']

# Calcular la frecuencia relativa
frecuencia['Frecuencia Relativa'] = frecuencia['Frecuencia Absoluta'] / 500

# Calcular la frecuencia acumulada
frecuencia['Frecuencia Acumulada'] = frecuencia['Frecuencia Absoluta'].cumsum()

# Ordenar de mayor a menor
frecuencia = frecuencia.sort_values(by='Frecuencia Absoluta', ascending=False).reset_index(drop=True)

coloresin = ['magenta', 'cyan', 'pink', 'purple', 'blue', 'green', 'yellow', 'orange', 'red']

# Colores para las barras
colores = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'cyan', 'magenta']

# Gráfica 1: Histograma de Frecuencia Absoluta
plt.figure(figsize=(10, 6))
plt.bar(frecuencia['Fruta'], frecuencia['Frecuencia Absoluta'], color=colores[:len(frecuencia)], width=0.99)
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfica 2: Diagrama de Barras Horizontales de Frecuencia Absoluta
plt.figure(figsize=(10, 6))
plt.barh(frecuencia['Fruta'][::-1], frecuencia['Frecuencia Absoluta'][::-1], color=coloresin[:len(frecuencia)], height=0.97)
plt.title('Diagrama de Barras Horizontales de Frecuencia Absoluta')
plt.xlabel('Frecuencia Absoluta')
plt.ylabel('Fruta')
plt.tight_layout()
plt.show()

# Gráfica 3: Polígono de Frecuencia (inicia y termina en cero)
plt.figure(figsize=(10, 6))
plt.plot(['Inicio'] + frecuencia['Fruta'].tolist() + ['Fin'], [0] + frecuencia['Frecuencia Absoluta'].tolist() + [0], marker='o')
plt.title('Polígono de Frecuencia')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.grid(True)
plt.fill_between(['Inicio'] + frecuencia['Fruta'].tolist() + ['Fin'], [0] + frecuencia['Frecuencia Absoluta'].tolist() + [0], alpha=0.1)
plt.tight_layout()
plt.show()

# Gráfica 4: Diagrama de Pastel de Frecuencia Relativa
plt.figure(figsize=(8, 8))
plt.pie(frecuencia['Frecuencia Relativa'], labels=frecuencia['Fruta'], autopct='%1.1f%%', startangle=90)
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Gráfica 5: Ojiva de Frecuencia Acumulada (inicia en cero)
plt.figure(figsize=(10, 6))
plt.plot(['Inicio'] + frecuencia['Fruta'].tolist(), [0] + frecuencia['Frecuencia Acumulada'].tolist(), marker='o', color='orange')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Fruta')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.fill_between(['Inicio'] + frecuencia['Fruta'].tolist(), [0] + frecuencia['Frecuencia Acumulada'].tolist(), alpha=0.1)
plt.show()
```
